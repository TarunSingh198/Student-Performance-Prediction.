```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
from io import StringIO

# Upload the CSV file
print("üìÅ Please upload your CSV file:")
uploaded = files.upload()

# Read the uploaded file
for filename in uploaded.keys():
    print(f"\n‚úÖ File '{filename}' uploaded successfully!\n")
    data = pd.read_csv(StringIO(uploaded[filename].decode('utf-8')))

# Display the first few rows
print("üîç Preview of the Dataset:")
display(data.head())

# Show basic information
print("\nüìå Dataset Info:")
print(data.info())

# Show basic statistics
print("\nüìä Statistical Summary:")
display(data.describe())

# Check for missing values
print("\n‚ùì Missing Values:")
print(data.isnull().sum())

# Visualizing distribution of scores (if columns like 'math score', 'reading score', 'writing score' exist)
score_columns = ['math score', 'reading score', 'writing score']
existing_scores = [col for col in score_columns if col in data.columns]

if existing_scores:
    print("\nüìà Score Distributions:")
    for col in existing_scores:
        plt.figure(figsize=(8, 4))
        sns.histplot(data[col], bins=20, kde=True, color='skyblue')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.show()

# Correlation heatmap
print("\nüß† Correlation Heatmap:")
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation between Numeric Features')
plt.show()

# Gender-based performance comparison (if 'gender' column exists)
if 'gender' in data.columns and existing_scores:
    print("\nüìä Gender-based Score Comparison:")
    for col in existing_scores:
        plt.figure(figsize=(6, 4))
        sns.boxplot(x='gender', y=col, data=data)
        plt.title(f'{col.capitalize()} by Gender')
        plt.show()

# Performance categorization based on average score (optional)
if len(existing_scores) == 3:
    data['average score'] = data[existing_scores].mean(axis=1)
    data['performance'] = pd.cut(data['average score'],
                                 bins=[0, 50, 70, 100],
                                 labels=['Low', 'Medium', 'High'])

    print("\nüéØ Performance Categories:")
    display(data[['average score', 'performance']].head())

    plt.figure(figsize=(6, 4))
    sns.countplot(x='performance', data=data, palette='Set2')
    plt.title('Performance Level Distribution')
    plt.show()
```
